```python
import matplotlib.pyplot as plt

lengths = df['DetalhamentoSolicitacao'].dropna().str.len()
plt.figure(figsize=(10, 3))
plt.boxplot(lengths, vert=False, showfliers=True)
plt.xlabel("Comprimento (número de caracteres)")
plt.title("Boxplot do comprimento de DetalhamentoSolicitacao")
plt.show()
```

```python
q1 = lengths.quantile(0.25)
q3 = lengths.quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = lengths[(lengths < lower) | (lengths > upper)]
print("Outliers (IQR):", len(outliers), "/", len(lengths), f"({len(outliers)/len(lengths):.4%})")
```

```python
mask = df['DetalhamentoSolicitacao'].str.len() < lower
indices = df.index[mask].tolist()
print(indices)
print(len(indices))
```

```python
mask = df['DetalhamentoSolicitacao'].str.len() > upper
indices = df.index[mask].tolist()
print(indices)
print(len(indices))
```
